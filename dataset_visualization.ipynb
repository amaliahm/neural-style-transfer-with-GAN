```python
import os
import random
import matplotlib.pyplot as plt
from PIL import Image

# Path to dataset (update this based on your local path)
dataset_path = "/kaggle/input/monet2photo"
monet_path = os.path.join(dataset_path, "trainA")  # Monet
photo_path = os.path.join(dataset_path, "trainB")  # Photo

def load_random_images(folder_path, num_images=5):
    files = os.listdir(folder_path)
    selected = random.sample(files, num_images)
    images = [Image.open(os.path.join(folder_path, file)).convert("RGB") for file in selected]
    return images, selected

def plot_images(images, titles, domain_name):
    plt.figure(figsize=(15, 3))
    for i, (img, title) in enumerate(zip(images, titles)):
        plt.subplot(1, len(images), i+1)
        plt.imshow(img)
        plt.title(f"{domain_name}\n{title}", fontsize=10)
        plt.axis("off")
    plt.tight_layout()
    plt.show()

# Load and display Monet images
monet_images, monet_titles = load_random_images(monet_path)
plot_images(monet_images, monet_titles, "Monet")

# Load and display Photo images
photo_images, photo_titles = load_random_images(photo_path)
plot_images(photo_images, photo_titles, "Photo")
```

```python
from collections import Counter

def analyze_dataset(folder_path):
    files = os.listdir(folder_path)
    print(f"Total images in {folder_path}: {len(files)}")
    dims = [Image.open(os.path.join(folder_path, file)).size for file in files[:100]]  # sample 100
    count = Counter(dims)
    print("Most common image sizes (in sample):", count.most_common(3))

analyze_dataset(monet_path)
analyze_dataset(photo_path)
```
